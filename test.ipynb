```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.karate_club_graph()

# Draw G, set to bigget figsize

plt.figure(figsize=(20,8))
nx.draw(G, with_labels=True, node_size=200, node_color="skyblue", alpha=0.7, linewidths=20)
```

```python
# partition with Kernighan Lin

from networkx.algorithms.community import kernighan_lin_bisection
comp = kernighan_lin_bisection(G) # returns a tuple of two sets of nodes

# add labels to graph

for n in G.nodes:
    G.nodes[n]["color"] = "skyblue"
    G.nodes[n]["pos"] = n

for n in comp[0]:
    G.nodes[n]["color"] = "red"
    G.nodes[n]["pos"] = n
    
for n in comp[1]:
    G.nodes[n]["color"] = "green"
    G.nodes[n]["pos"] = n

# Draw the partitioned graph

plt.figure(figsize=(20,8))
pos = nx.spring_layout(G, seed=42)
colors = [G.nodes[n]["color"] for n in G.nodes]
nx.draw(G, pos, with_labels=True, node_size=200, node_color=colors, alpha=0.7, linewidths=20)
```
